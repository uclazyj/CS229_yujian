(a)
$$f(x) = \frac{1}{2}x^T A x + b^T x = \frac{1}{2} \sum_{i=1}^n \sum_{j=1}^n x_i A_{ij} x_j + \sum_{i=1}^n b_i x_i$$
$$
\begin{align}
\frac{\partial}{\partial x_k} f(x) 
&= \frac{1}{2} \sum_{i=1}^n \sum_{j=1}^n (\delta_{ik} A_{ij} x_j + x_i A_{ij} \delta_{jk}) + \sum_{i=1}^n b_i \delta_{ik}
\\
&= \frac{1}{2} (\sum_{j=1}^n A_{kj} x_j + \sum_{i=1}^n x_i A_{ik}) + b_k
\\
&= \frac{1}{2} (\sum_{j=1}^n A_{kj} x_j + \sum_{i=1}^n A_{ki} x_i) + b_k
\\
&= \frac{1}{2} (\sum_{j=1}^n A_{kj} x_j + \sum_{i=1}^n A_{ki} x_i) + b_k
\\
&=\sum_{i=1}^n A_{ki} x_i + b_k
\\
&=(A x)_k + b_k
\end{align}
$$
So 
$$\nabla f(x) = Ax + b$$

(c)

$$\frac{\partial^2}{\partial x_i \partial x_j} f(x) = \frac{1}{2} \frac{\partial^2}{\partial x_i \partial x_j} \sum_{k=1}^n \sum_{l=1}^n x_k A_{kl} x_l$$

Only two terms in the sum have contributions: $k=i, l=j$ and $k=j,l=i$. So

$$\frac{\partial^2}{\partial x_i \partial x_j} f(x) = \frac{1}{2} \frac{\partial^2}{\partial x_i \partial x_j} (x_i A_{ij} x_j + x_j A_{ji} x_i)  = \frac{1}{2}(A_{ij} + A_{ji})$$

Since $A$ is symmetric, so $A_{ij} = A_{ji}$. So:

$$\frac{\partial^2}{\partial x_i \partial x_j} f(x) =A_{ij}$$

So

$$\nabla^2 f(x) = A$$

(d)
$$\frac{\partial}{\partial x_i} f(x) = \frac{\partial}{\partial x_i} g(a^T x) = g'(a^T x)\frac{\partial}{\partial x_i} \sum_{j=1}^n (a_j x_j)= g'(a^T x) \sum_{j=1}^n a_j \delta_{ij} = g'(a^T x) a_i$$
So
$$\nabla f(x) = g'(a^T x) a$$

Then we calculate $\nabla^2 f(x)$. The result is a matrix, so let's calculate the element on $i_{th}$ row and $j_{th}$ column.

$$
\begin{align}
(\nabla^2 f(x))_{ij} &= \frac{\partial^2}{\partial x_i \partial x_j} g(a^T x)
\\
&= \frac{\partial}{\partial x_i} g'(a^T x) a_j
\\
&= a_i a_j g''(a^T x)
\end{align}
$$

Therefore

$$\nabla^2 f(x) = aa^T g''(a^T x)$$

(a)

For any vector $x$, we have

$$x^T A x = x^T zz^T x = (z^T x)^T z^T x \geq 0$$

Therefore $A$ is positive semidefinite.

(b)

If $x$ is in the null-space of $A$, we must have: $A x = 0$, which means $zz^T x = 0$. Therefore, $x^T zz^T x= 0$. So $(z^T x)^Tz^T x= 0$. So $z^T x = 0$.

So the null-space of $A$ is a set of all vectors $x$ such that $z^T x = 0$.

It is easy to see the rank of $A$ is at most 1, because any two rows are linearly dependent: The $i_{th}$ row is $z_i z^T$, and the $j_{th}$ row is $z_j z^T$.

Since $z$ is a non-zero vector, there is at least one element in the diagonal one $A$ that is not zero. So the rank of $A$ is at least 1.

Therefore, the rank of $A$ is 1.
