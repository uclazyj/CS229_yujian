(a)

$$p(y;\lambda) = \frac{e^{-\lambda} \lambda^y}{y!} = \frac{e^{-\lambda} e^{y\ln\lambda}}{y!} = \frac{e^{y\ln\lambda-\lambda}}{y!}$$

let $\eta = \ln \lambda$, we have

$$p(y;\eta) = \frac{1}{y!}\exp(\eta y - e^{\eta})$$

Compare with the expression of exponential family distribution

$$p(y;\eta) =b(y)\exp(\eta^T T(y) - a(\eta))$$

We have 
$$b(y) = \frac{1}{y!}$$

$$\eta = \ln \lambda$$

$$T(y) = y$$

$$a(\eta) = e^{\eta}$$

(b)

$$h_\theta(x) = E[y|x; \theta] = \lambda = e^\eta = e^{\theta^T x}$$

(c)

$$l = \ln \prod_{i=1}^m p^{(i)} = \sum_{i=1}^m \ln p^{(i)} = \sum_{i=1}^m \ln \frac{e^{-\lambda} \lambda^{y^{(i)}}}{y^{(i)}!} = \sum_{i=1}^m y^{(i)} \ln \lambda - \lambda - \ln y^{(i)}!$$

$$\frac{\partial l}{\partial \theta_j} = \frac{\partial l}{\partial \lambda}\frac{\partial \lambda}{\partial \theta_j} = \sum_{i=1}^m (\frac{y^{(i)}}{\lambda} - 1) e^{\theta^T x^{(i)}} x_j^{(i)} = \sum_{i=1}^m (y^{(i)} - e^{\theta^T x^{(i)}}) x_j^{(i)}  = \sum_{i=1}^m (y^{(i)} - h_\theta(x^{(i)})) x_j^{(i)}$$

So the stochastic gradient ascent update rule is:

$$\theta_j := \theta_j + \alpha \frac{\partial l^{(i)}}{\partial \theta_j} = \theta_j + \alpha (y^{(i)} - h_\theta(x)) x_j^{(i)}$$

(d)

For implementation convenience, we write the gradient ascent rule in a matrix form:

$$\theta := \theta + \alpha X^T (Y - e^{X\theta})$$

Note that choosing an appropriate learning rate $\alpha$ is essential for training the model. If the learning rate is too big, then we will encounter the overflow error.
