(a)

According to the chain of probablity, we have

$$p(y^{(i)} = 1, t^{(i)} = 1, x^{(i)}) = p(x^{(i)}) p(t^{(i)} = 1 | x^{(i)}) p(y^{(i)}=1|t^{(i)} = 1 , x^{(i)})$$

and

$$p(y^{(i)} = 1, t^{(i)} = 1, x^{(i)}) = p(x^{(i)}) p(y^{(i)} = 1 | x^{(i)}) p(t^{(i)}=1|y^{(i)} = 1 , x^{(i)})$$

so

$$p(t^{(i)} = 1 | x^{(i)}) p(y^{(i)}=1|t^{(i)} = 1 , x^{(i)}) = p(y^{(i)} = 1 | x^{(i)}) p(t^{(i)}=1|y^{(i)} = 1 , x^{(i)})$$

Since $p(y^{(i)}=1|t^{(i)} = 1 , x^{(i)}) = p(y^{(i)}=1|t^{(i)} = 1)$ and $p(t^{(i)}=1|y^{(i)} = 1)=1$, we have

$$p(t^{(i)} = 1 | x^{(i)}) p(y^{(i)}=1|t^{(i)} = 1) = p(y^{(i)} = 1 | x^{(i)}) $$


So 

$$p(t^{(i)} = 1 | x^{(i)})  = p(y^{(i)} = 1 | x^{(i)}) / \alpha$$

where $\alpha = p(y^{(i)}=1|t^{(i)} = 1)$, which means the probablity of a positive data being labeled.

Update: After being confused for a long time, I think I finally figure it out: The model $h(x)$, which hopefully approximates $p(y=1 | x)$, is trained using all the data points $x^{(i)}$. Then, we can always pick a specific $x^{(i)}$ from $V_{+}$, and the model predicts the probability of the label being 1. From $p(y^{(i)} = 1 | t^{(i)} = 1,x^{(i)}) = p(y^{(i)} = 1 | t^{(i)} = 1)$ we know that given a positive data, we can never tell whether it is labeled ($y^{(i)}=1$) based on the data $x^{(i)}$: A data point $x^{(i)}$ with label $y^{(i)} = 0$ but $t^{(i)} = 1$ should look extremely similar. We just cannot tell. So $p(y^{(i)} = 1 | x^{(i)})$ cannot be 1, unless all the positive data points are labeled (then we can conclude confidently that the given data point is labeled).

In other words, even if we know we pick $x^{(i)}$ from $V_+$, the model $h(x)$ cannot take advantage of that, due to the assumption $p(y^{(i)} = 1 | t^{(i)} = 1,x^{(i)}) = p(y^{(i)} = 1 | t^{(i)} = 1)$.
