# 5. Kernelizing the Perceptron

(a)

i.

$$\theta^{(i+1)}:=\theta^{(i)} + \alpha(y^{(i+1)} - h_{\theta^{(i)}}(x^{(i+1)}))x^{(i+1)} = \theta^{(i)} + \alpha[y^{(i+1)} - g(\theta^{(i)T}x^{(i+1)})]x^{(i+1)}$$

Now we map $x$ to $\phi(x)$, then $\theta$ will be the dimension of $\phi(x)$, rather than the dimension of $x$. The update rule becomes:

$$\theta^{(i+1)}:= \theta^{(i)} + \alpha[y^{(i+1)} - g(\theta^{(i)T}\phi(x^{(i+1)}))]\phi(x^{(i+1)})$$

$\theta^{(0)}$ is still a zero vector, but with a dimension that is same as the dimension of $\phi(x)$.

It is easy to notice that $\theta^{(1)}$ is proportional to $\phi(x^{(1)})$, $\theta^{(2)}$ is a linear combination of $\phi(x^{(1)})$ and $\phi(x^{(2)})$, $\theta^{(3)}$ is a linear combination of $\phi(x^{(1)})$, $\phi(x^{(2)})$ and $\phi(x^{(3)})$, etc. Therefore, we can represent $\theta^{(i)}$ as a linear combination of $\phi(x^{(1)})$, $\phi(x^{(2)})$, ... ,$\phi(x^{(i)})$:

$$\theta^{(i)} = \sum_{j=1}^i \beta_j \phi(x^{(j)})$$

ii.

$$
\begin{align}
h_{\theta^{(i)}}(x^{(i+1)}) &= g\left( \theta^{(i)T} \phi(x^{(i+1)}) \right)
\\
&= g\left(\sum_{j=1}^i \beta_j \phi(x^{(j)})^T \phi(x^{(i+1)}) \right)
\\
&= g\left(\sum_{j=1}^i \beta_j \langle \phi(x^{(j)}), \phi(x^{(i+1)}) \rangle \right)
\\
&= g\left(\sum_{j=1}^i \beta_j K(x^{(j)}, x^{(i+1)})\right)
\end{align}
$$

iii.

$$
\begin{align}
\theta^{(i+1)}&:= \sum_{j=1}^i \beta_j \phi(x^{(j)}) + \alpha \left[y^{(i+1)} - g\left(\sum_{j=1}^i \beta_j K(x^{(j)}, x^{(i+1)})\right) \right]\phi(x^{(i+1)})
\\
&=\sum_{j=1}^{i+1} \beta_j \phi(x^{(j)})
\end{align}$$

where

$$\beta_{i+1} = \alpha \left[y^{(i+1)} - g\left(\sum_{j=1}^i \beta_j K(x^{(j)}, x^{(i+1)})\right) \right]$$

(b)

We need to determine the value of $\beta_1$.

$$\theta^{(1)} = \theta^{(0)} + \alpha[y^{(1)} - g(\theta^{(0)T}\phi(x^{(1)}))]\phi(x^{(1)}) = \alpha(y^{(1)} - 1)\phi(x^{(1)})$$

Therefore 

$$\beta_1 = \alpha(y^{(1)} - 1)$$

At prediction time, we do

$$g(\theta^T \phi(x)) = g(\theta^{(m)T} \phi(x)) = g\left(\sum_{j=1}^m \beta_j \phi(x^{(j)})\phi(x)\right) = g\left(\sum_{j=1}^m \beta_j K(x^{(j)},x)\right)$$
