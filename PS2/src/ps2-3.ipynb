(a)

We know that

$$p(\theta|x,y)p(y|x)p(x)=p(x,y,\theta) = p(y|x,\theta)p(\theta|x)p(x)$$

So

$$p(\theta|x,y) = \frac{p(y|x,\theta)p(\theta|x)}{p(y|x)}$$

Therefore,

$$
\begin{align}
\theta_{MAP} &= \arg \max_\theta p(\theta|x,y)
\\
&= \arg \max_\theta \frac{p(y|x,\theta)p(\theta|x)}{p(y|x)}
\\
&= \arg \max_\theta p(y|x,\theta)p(\theta|x)
\\
&= \arg \max_\theta p(y|x,\theta)p(\theta)
\end{align}
$$

(b)

Start from the result in (a), we have

$$
\begin{align}
\theta_{MAP}
&= \arg \max_\theta p(y|x,\theta)p(\theta)
\\
&= \arg \max_\theta \ln[p(y|x,\theta)p(\theta)]
\\
&= \arg \min_\theta -\ln[p(y|x,\theta)p(\theta)]
\\
&= \arg \min_\theta -\ln p(y|x,\theta) - \ln p(\theta)
\\
&= \arg \min_\theta -\ln p(y|x,\theta) - \ln \frac{1}{(2\pi)^{n/2}\eta} e^{-\frac{||\theta||_2^2}{2\eta^2}}
\\
&= \arg \min_\theta -\ln p(y|x,\theta) +\frac{||\theta||_2^2}{2\eta^2}
\end{align}
$$

so

$$\lambda = \frac{1}{2\eta^2}$$

(c)

We know that 

$$p(y|x,\theta) \sim N(\theta^T x, \sigma^2) = \frac{1}{\sqrt{2\pi}\sigma}\exp[-\frac{(y-\theta^Tx)^2}{2\sigma^2}]$$

Plug in the result of (b), we have

$$
\begin{align}
\theta_{MAP} 
&= \arg \min_\theta \sum_{i=1}^m \frac{(y^{(i)}-\theta^Tx^{(i)})^2}{2\sigma^2} +\lambda||\theta||_2^2
\\
&= \arg \min_\theta \frac{1}{2\sigma^2} (X\theta -y)^T(X\theta -y) +\frac{1}{2\eta^2} \theta^T \theta
\end{align}$$

The way to calculate $\theta_{MAP}$ is to take the gradient with respect to $\theta$ and set the gradient to 0.

$$\nabla_\theta [\frac{1}{2\sigma^2} (X\theta -y)^T(X\theta -y) +\frac{1}{2\eta^2} \theta^T \theta] = \frac{1}{\sigma^2}(X^T X \theta - X^T y) + \frac{1}{\eta^2} \theta$$

Setting the derivative above to 0, we can solve for $\theta$, which is just $\theta_{MAP}$:

$$\theta_{MAP} = (X^TX +\frac{\sigma^2}{\eta^2} I)^{-1}X^Ty$$

(d)

We start from the result in (a)

$$\theta_{MAP} =\arg \max_\theta p(y|x,\theta)p(\theta)$$

Or 

$$\theta_{MAP} =\arg \max_\theta \ln (p(y|x,\theta)p(\theta)) = \arg \min_\theta (-\ln p(y|x,\theta) -\ln p(\theta))$$

Then we plug in

$$p(y|x,\theta) \sim N(\theta^T x, \sigma^2) = \frac{1}{\sqrt{2\pi}\sigma}\exp[-\frac{(y-\theta^Tx)^2}{2\sigma^2}]$$

and the Laplace prior

$$p(\theta) = \frac{1}{(2b)^m} \exp(-\frac{||\theta||_1}{b})$$

We then have

$$
\begin{align}
\theta_{MAP} &=\arg \min_\theta (-\ln p(y|x,\theta) -\ln p(\theta))
\\
&=\arg \min_\theta (\sum_{i=1}^m \frac{(y^{(i)}-\theta^Tx^{(i)})^2}{2\sigma^2} + \frac{||\theta||_1}{b})
\\
&=\arg \min_\theta (\frac{||X\theta - y||_2^2}{2\sigma^2} + \frac{||\theta||_1}{b})
\\
&=\arg \min_\theta (||X\theta - y||_2^2 +  \frac{2\sigma^2}{b}||\theta||_1)
\\
&=\arg \min_\theta J(\theta)
\end{align}$$

where
$$J(\theta) = ||X\theta - y||_2^2 +  \frac{2\sigma^2}{b}||\theta||_1 $$

We can see that 

$$\gamma = \frac{2\sigma^2}{b}$$
