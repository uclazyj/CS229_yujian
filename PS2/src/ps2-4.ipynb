(a)

Yes. 

The sum of two positive semidefinite (PSD) matrices is still a PSD matrix.

(b)

No. 

Counter example: $\phi_1(x) = x, \phi_2(x) = 2x$. Then 
$$K(x,z) = K_1(x,z)-K_2(x,z) = \phi_1(x)^T\phi_1(z) - \phi_2(x)^T\phi_2(z) = -3x^Tz$$

So for a non zero training example $x^{(i)}$

$$K_{ii} = K(x^{(i)},x^{(i)}) = -3(x^{(i)})^T x^{(i)} < 0$$

So $K$ is not a PSD matrix. Therefore, $K(x,z)$ is not a kernel.

(e)

Yes.

$$
K_{ij} = K(x^{(i)},x^{(j)}) = K_1(x^{(i)},x^{(j)}) K_2(x^{(i)},x^{(j)}) =\phi_1(x^{(i)})^T\phi_1(x^{(j)})\phi_2(x^{(i)})^T\phi_2(x^{(j)})$$

For any vector $z$, we have

$$
\begin{align}
z^T K z &= \sum_{i=1}^n \sum_{j=1}^n z_i K_{ij} z_j 
\\
&= \sum_{i=1}^n \sum_{j=1}^n z_i \phi_1(x^{(i)})^T\phi_1(x^{(j)})\phi_2(x^{(i)})^T\phi_2(x^{(j)}) z_j
\\
&= \sum_{i=1}^n \sum_{j=1}^n z_i \sum_{k=1}^d\phi_{1k}(x^{(i)})\phi_{1k}(x^{(j)}) \sum_{l=1}^d \phi_{2l}(x^{(i)})\phi_{2l}(x^{(j)}) z_j
\\
&= \sum_{i=1}^n \sum_{j=1}^n \sum_{k=1}^d \sum_{l=1}^d z_i \phi_{1k}(x^{(i)})\phi_{1k}(x^{(j)}) \phi_{2l}(x^{(i)})\phi_{2l}(x^{(j)}) z_j
\\
&= \sum_{k=1}^d \sum_{l=1}^d \sum_{i=1}^n z_i \phi_{1k}(x^{(i)}) \phi_{2l}(x^{(i)}) \sum_{j=1}^n z_j\phi_{1k}(x^{(j)}) \phi_{2l}(x^{(j)})
\\
&= \sum_{k=1}^d \sum_{l=1}^d (\sum_{i=1}^n z_i \phi_{1k}(x^{(i)}) \phi_{2l}(x^{(i)}))^2
\\
&\geq 0
\end{align}
$$

(f)

Yes.

$$K_{ij} = K(x^{(i)},x^{(j)}) = f(x^{(i)})f(x^{(j)})$$

For any vector $z$, we have

$$z^T K z = \sum_i \sum_j z_i K_{ij} z_j = \sum_i \sum_j z_i f(x^{(i)})f(x^{(j)}) z_j = \sum_i  z_i f(x^{(i)})\sum_j z_j f(x^{(j)}) = (\sum_i  z_i f(x^{(i)}))^2 \geq 0$$

$$
